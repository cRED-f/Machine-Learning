# Manipulating data 2

```python
car_sales_suffled=car_sales.sample(frac=1)
# here take car_sales table, then take sample(car_sales table),
# frac (fraction of sample) ex: frac=0.5 (meaning take 5 row out of 10 row(sample) randomly) ex: frac=1 (take all the data)
```

```python
car_sales_suffled
```

# Note:On real world problem when data contain millions of rows then only select 20% of data

```python
car_sales_suffled.sample(frac=0.2)
```

```python
# Reset the row indexes order
car_sales_suffled.reset_index(drop=True,inplace=True) # drop=True will delete extra index values
```
