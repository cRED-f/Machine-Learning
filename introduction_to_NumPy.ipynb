# Creating Arrays

```python
sample_arr=np.array([1,2,3])
sample_arr
```

```python
sample_arr.dtype
```

```python
# Return a new array of given shape and type, filled with ones.
ones=np.ones((3,3))
ones
```

```python
ones.dtype
```

```python
type(ones)
```

```python
# Return a new array of given shape and type, filled with zeros.
zeros=np.zeros((3,3))
zeros
```

```python
zeros.dtype
```

```python
# Return evenly spaced values within a given interval.
#arange(start,stoped,step)
range_array = np.arange(0,10,2)
range_array
```

```python
range_array.dtype
```

```python
# Return random integers from low (inclusive) to high' (exclusive).
# randint(low , high=None , size=None)
random_array = np.random.randint(0,10,size=(3,5))
random_array
```

```python
random_array.dtype
```

```python
random_array.shape
```

```python
random_array.size
```

```python
# another way to create random array
random_array2=np.random.random((2,3))
random_array2
```

```python
random_array2.shape
```

```python
# another way to create random array
random_array3=np.random.rand(2,3)
random_array3
```

# Pseudo-random numbers
# np.random.seed()

```python
random_array4=np.random.randint(0,10,size=(2,3))
random_array4
#values will change after refreshes
```

```python
# if we want to stop this we can use seed() fucntion
np.random.seed(seed=0)
random_array4=np.random.randint(0,10,size=(2,3))
random_array4
```
