# 1st method

```python
fig = plt.figure() #create a figure
ax = fig.add_subplot() # adds some axes
plt.show()
```

# 2nd method

```python
fig = plt.figure() # create a figure
ax= fig.add_axes([1,1,1,1])
ax.plot(x,y) # add some data
plt.show()
```

# 3rd method (recommended)

```python
fig,ax =plt.subplots()
ax.plot(x,y); #add some data
```

# Matplotlib example workflow

Screenshot 2023-12-04 at 9.54.10 PM.png

```python
# 0.import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1.prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# 2. setup plot (object-oriented method)
fig,ax = plt.subplots(figsize=(10,10)) # here figsize= width , height

# 3. plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save and show (you save the whole figure)
fig.savefig("sample-plot.png")
```
