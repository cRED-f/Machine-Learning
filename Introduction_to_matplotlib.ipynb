```python
# 0.import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1.prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# 2. setup plot (object-oriented method)
fig,ax = plt.subplots(figsize=(10,10)) # here figsize= width , height

# 3. plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save and show (you save the whole figure)
fig.savefig("sample-plot.png")
```

# Bar plot


* You can create vertical bar plots with ax.bar() and horizontal bar plots with ax.barh().

# Vertical bar plot

```python
# make a plot from python dictionay
nut_butter_prices={
    "Somosa":10,
    "Puri":5,
    "Bel Puri":8
}
fig,ax=plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values()); #x(width),y(height)
ax.set(title="Fahim's nut butter store",
       ylabel="Price ($)")
```

# horizontal bar plot

```python
fig,ax=plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
ax.set(title="Fahim's nut butter store",
       xlabel="Price ($)")
```

# Subplots

- Subplots allow you to create multiple Axes on the same Figure (multiple plots within the same plot).

- Subplots are helpful because you start with one plot per Figure but scale it up to more when necessary.

- For example, let's create a subplot that shows many of the above datasets on the same Figure.

- We can do so by creating multiple Axes with plt.subplots() and setting the nrows (number of rows) and ncols (number of columns) parameters to reflect how many Axes we'd like.

- nrows and ncols parameters are multiplicative, meaning plt.subplots(nrows=2, ncols=2) will create 2*2=4 total Axes.



# 1st method for Subplot

```python
# Option 1: Create 4 subplots with each Axes having its own variable name
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# plot data to each axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

# 2nd method for Subplot

```python
fix,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# plot to each different index
ax[0,0].plot(x,x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

# which should I use? (pyplot vs matplotlib oo method)

* when plotting something quickly, we should use pyplot method
* when plotting something more advanced, we should use oo method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
len(over_50)
```

```python
over_50.head()
```

#pyplot method

```python
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

# oo method mixed with pyplot method

```python
fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
```

```python
# create limitation in x axis
fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
ax.set_xlim([45,100]);
```

# oo method from scratch(object-oriented method)

```python
fig,ax=plt.subplots(figsize=(10,6))

# plot the data
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"]);


# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
over_50.target.values
```

# Subplot heart disease(oo method)

```python
# subplot of chol, age, thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True);
# plot the ax0
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"]);


# customize the plot ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
      # xlabel="Age", #(because sharex=True)
       ylabel="Cholesterol");

# add a legend ax0
ax0.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line ax0
ax0.axhline(over_50["chol"].mean(),linestyle="--");



# plot the ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"]);


# customize the plot ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
       xlabel="Age",
       ylabel="Max Heart Rate");

# add a legend ax1
ax1.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line ax1
ax1.axhline(over_50["thalach"].mean(),linestyle="--");


#add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");

#Save and show (you save the whole figure)
#fig.savefig("new-plot.png")
```

# Customizing the y and x axis limitations

```python
# subplot of chol, age, thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True);
# plot the ax0
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                     cmap="winter"); #this changes the colour scheme ("winter/summer")


# change the x axis limits
ax0.set_xlim([50,80])


# customize the plot ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
      # xlabel="Age", #(because sharex=True)
       ylabel="Cholesterol");

# add a legend ax0
ax0.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line ax0
ax0.axhline(over_50["chol"].mean(),linestyle="--");



# plot the ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
              cmap="winter"); #this changes the colour scheme ("winter/summer")


# change the x axis limits
ax1.set_xlim([50,80])


# change the y axis limits
ax1.set_ylim([60,200])


# customize the plot ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
       xlabel="Age",
       ylabel="Max Heart Rate");

# add a legend ax1
ax1.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line ax1
ax1.axhline(over_50["thalach"].mean(),linestyle="--");


#add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");

#Save and show (you save the whole figure)
#fig.savefig("new-plot.png")
```
